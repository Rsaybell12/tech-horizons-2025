```python
import pandas as pd
datos = pd.read_csv("datosclientes.csv")
print(datos.head())
print(f"Filas: {datos.shape[0]}, Columnas: {datos.shape[1]}")
datos = datos.dropna() #eliminar datos nulos
```

```python
# ANALISIS EXPLORATORIO DE DATOS
datos.info()
```

```python
datos.describe()
# count = significa el numero de filas
# mean = el promedio de los valores
# std = aber si los datos son parecidos o muy diferentes
# insights : "Los códigos empiezan en 1 y terminan en 2709
# significa que no hay saltos grandes, no hay códigos negativos y no hay valores fuera de rango"
```

```python
# Filtro 1: clientes con saldo mayor a 1000
filtro1 = datos[datos["saldo"] > 1000]
print("\n--- Filtro 1: Saldo mayor a 1000 ---")
print(filtro1)

# Filtro 2: clientes con debito mayor a 500
filtro2 = datos[datos["debito"] > 500]
print("\n--- Filtro 2: Débito mayor a 500 ---")
print(filtro2)

# Filtro 3: clientes con credito mayor a 0
filtro3 = datos[datos["credito"] > 0]
print("\n--- Filtro 3: Crédito mayor a 0 ---")
print(filtro2)

```

```python
# Crear columna saldo_final
datos["saldo_final"] = datos["saldo"] + datos["credito"] - datos["debito"]

# Crear columna categoria_saldo
datos["categoria_saldo"] = "Bajo"   # primero asignamos lo mismo a todos

# Luego cambiamos condiciones básicas
datos.loc[datos["saldo"] > 500, "categoria_saldo"] = "Medio"
datos.loc[datos["saldo"] > 2000, "categoria_saldo"] = "Alto" # el loc...sirve para elegir filas con una condición y cambiar o ver solo esas partes del DataFrame.
print(datos.head())
```

```python
# Promedio de saldo por categoría
promedio_saldo = datos.groupby("categoria_saldo")["saldo"].mean()
print(promedio_saldo) # calcula el prodmedio del saldo segun la categoria
                       # osea agrupa los datos por categoria y obtiene el saldo
                       #promedio de cada grupo

# Conteo por dirección (direcc1)
clientes_por_direccion = datos.groupby("direcc1")["codigo"].count()
print(clientes_por_direccion) #cuenta cuantos clientes hay en cada direccion,
                               #agrupando por la columna de direccion y mostrando cuantos registros pertencen a cada uno.
```
