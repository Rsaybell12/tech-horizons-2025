#Primer Grafico

```python
import pandas as pd
import matplotlib.pyplot as plt

# Asegurar formato fecha
data["post_time"] = pd.to_datetime(data["post_time"])
#agrupar likes por mes
likes_mes = data.resample("ME", on="post_time")["likes"].sum()
#crearlo con diferente estilo
plt.plot(likes_mes.index, likes_mes.values, 'g^:', label='Likes')
#titulo y etiquetas
plt.title("Likes por Mes")
plt.xlabel("Mes")
plt.ylabel("Total de Likes")
plt.legend()
# agregar cuadricula
plt.grid(True)
plt.savefig('grafico.png', dpi=300) # 300 DPI #guardar en formato imagen
plt.show()

#Este gráfico muestra cómo cambiaron
#los likes totales por mes. Permite ver si las interacciones aumentaron,
#bajaron o se mantuvieron estables con el tiempo.
#HISTORIA DEL PRIMER GRAFICO
```

#Segundo Grafico

```python
# Agrupar comentarios por mes
comentarios_mes = data.resample("ME", on="post_time")["comments"].sum()

plt.plot(comentarios_mes.index, comentarios_mes.values, 'ro--', label='Comentarios')

plt.title("Comentarios por Mes")
plt.xlabel("Mes")
plt.ylabel("Total de Comentarios")
plt.legend()
plt.grid(True)
plt.savefig('grafico.png', dpi=300) # 300 DPI
plt.show()

#HISTORIA
#El gráfico muestra cómo cambiaron los comentarios cada mes y permite ver cuándo hubo más o menos interacción.
```

#Tercer Grafico


```python
# Agrupar likes y shares por mes
likes_mes = data.resample("ME", on="post_time")["likes"].sum()
shares_mes = data.resample("ME", on="post_time")["shares"].sum()

plt.plot(likes_mes.index, likes_mes.values, 'k*-.', label='Likes')
plt.plot(shares_mes.index, shares_mes.values, 'bs--', label='Shares')

plt.title("Comparación: Likes vs Shares por Mes")
plt.xlabel("Mes")
plt.ylabel("Total Interacciones")
plt.legend()
plt.grid(True)
plt.savefig('grafico.png', dpi=300) # 300 DPI
plt.show()

#Historia
#Este gráfico permite comparar si los usuarios reaccionan más dando likes o compartiendo publicaciones.
```

```python
# Promedio de shares por tipo de publicación
shares_tipo = data.groupby("post_type")["shares"].mean()

# Graficar
plt.plot(shares_tipo.index, shares_tipo.values, 'g^-', label="Shares")  # línea verde con triángulos
plt.title("Promedio de Shares por Tipo de Publicación")
plt.xlabel("Tipo de Publicación")
plt.ylabel("Shares Promedio")
plt.grid(True)
plt.legend()
plt.savefig('grafico.png', dpi=300) # 300 DPI
plt.show()

#HISTORIA
#Muestra qué tipo de publicación recibe más compartidos en promedio
```
