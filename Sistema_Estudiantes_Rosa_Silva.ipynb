```python
# SISTEMA DE ESTUDIANTES
estudiantes = []

#  agregar estudiante
def agregar_estudiante(cedula, nombres, apellidos, edad, direccion, grado, materias_favoritas, promedio):
  estudiante =  {
      "cedula": cedula,
      "nombres": nombres,
      "apellidos": apellidos,
      "edad": edad,
      "direccion": direccion,
      "grado": grado,
      "materias_favoritas": materias_favoritas,
      "promedio": promedio
  }
  estudiantes.append(estudiante)

# buscar un estudiante por su nombre
def buscar_estudiante_nombre(nombre):
  for estudiante in estudiantes:
    if estudiante["nombres"] == nombre:
      return estudiante
  return None

#  buscar por su cédula
def buscar_estudiante_cedula(cedula):
  for estudiante in estudiantes:
    if estudiante["cedula"] == cedula:
      return estudiante
  return None

#  calcular el promedio
def calcular_promedio_general():
  if not estudiantes:
    return 0
  total = sum(est["promedio"] for est in estudiantes)
  return total / len(estudiantes)

#  mostrar todos los estudiantes
def mostrar_estudiantes():
  print("====Lista estudiantes====")
  for est in estudiantes:
    print(f"Cedula: {est['cedula']}")
    print(f"Nombres: {est['nombres']}")
    print(f"Apellidos: {est['apellidos']}")
    print(f"Edad: {est['edad']}")
    print(f"Direccion: {est['direccion']}")
    print(f"Grado: {est['grado']}")
    print(f"Materias favoritas: {', '.join(est['materias_favoritas'])}")
    print(f"Promedio: {est['promedio']}")
    print("------------------------")

# eliminar el estudiante
def eliminar_estudiante(cedula):
   for estudiante in estudiantes:
       if estudiante["cedula"] == cedula:
        estudiantes.remove(estudiante)
        return f"estudiante con cedula {cedula} eliminado"
   return "no encontrado"

#  actualizar la información
def actualizar_estudiante(cedula, nuevos_datos):
   for estudiante in estudiantes:
    if estudiante["cedula"] == cedula:
      estudiante.update(nuevos_datos)
      return f"estudiante con cedula {cedula} actualizado"
   return "no encontrado"

# Menuu
def menu():
    while True:
       print("elije una opcion")
       print("1. agrega estudiante")
       print("2. Buscar por nombre")
       print("3. Buscar por cedula")
       print("4. Calcular promedio")
       print("5. Mostrar todos los estudiantes")
       print("6. elimina estudiante")
       print("7. actualizar estudiante")
       print("8. Salir")

       opcion = input("pon una opcion:  ")

       if opcion == "1":
         cedula = input("cedula: ")
         nombres = input("nombres: ")
         apellidos = input("apellidos: ")
         edad = int(input("edad: "))
         direccion = input("direccion: ")
         grado = input("grado: ")
         materias_favoritas = input("materias favoritas: ").split(",")
         promedio = float(input("promedio: "))
         agregar_estudiante(cedula, nombres, apellidos, edad, direccion, grado, materias_favoritas, promedio)

       elif opcion == "2":
          nombre = input("nombre del estudiante:  ")
          estudiante = buscar_estudiante_nombre(nombre)
          if estudiante:
            print(estudiante)
          else:
            print("no encontrado")

       elif opcion == "3":
          cedula = input("cedula del estudiante:  ")
          estudiante = buscar_estudiante_cedula(cedula)
          if estudiante:
            print(estudiante)
          else:
            print("no encontrado")
       elif opcion == "4":
          promedio_general = calcular_promedio_general()
          print(f"el promedio general es: {promedio_general:.2f}")

       elif opcion == "5":
          mostrar_estudiantes()

       elif opcion == "6":
          cedula = input("cedula del estudiante: ")
          print(eliminar_estudiante(cedula))

       elif opcion == "7":
          cedula = input("cedula del estudiante: ")
          nuevos_datos = {}
          nuevo_grado = input("nuevo grado: ")
          if nuevo_grado:
            nuevos_datos["grado"] = nuevo_grado
          nueva_materias = input("nuevas materias: ").split(",")
          if nueva_materias:
            nuevos_datos["materias_favoritas"] = nueva_materias
          nuevo_promedio = input("nuevo promedio: ")
          if nuevo_promedio:
            nuevos_datos["promedio"] = float(nuevo_promedio)
          if edad:
            nuevos_datos["edad"] = int(edad)
          print(actualizar_estudiante(cedula, nuevos_datos))

       elif opcion == "8":
         break
       else:
          print("opcion no valida")
# ejecuta el menu
menu()
```
